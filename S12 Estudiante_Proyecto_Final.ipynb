```python
# 1. Unir órdenes con catálogo para obtener costos unitarios
# Usamos 'left' para no perder órdenes si algún producto no estuviera en el catálogo
orders_with_costs = orders_clean.merge(catalog_clean[['nombre_producto', 'costo_unitario']], 
                                       on='nombre_producto', 
                                       how='left')

# 2. Calcular el costo total por línea de pedido
orders_with_costs['costo_total_pedido'] = orders_with_costs['cantidad'] * orders_with_costs['costo_unitario']

# 3. Calcular el profit (margen) por línea de pedido: Revenue - Costos
# Nota: monto_total ya descuenta los 'monto_descuento' según la exploración previa
orders_with_costs['profit_pedido'] = orders_with_costs['monto_total'] - orders_with_costs['costo_total_pedido']

print("✅ Datos consolidados y márgenes calculados por pedido.")
```

```python
# Cálculos globales
total_revenue = orders_with_costs['monto_total'].sum()
total_cost_goods = orders_with_costs['costo_total_pedido'].sum()
total_marketing = marketing_clean['gasto'].sum()
net_profit = total_revenue - total_cost_goods - total_marketing

# Mostrar resultados con formato ejecutivo
print(f"--- Rentabilidad Global RappiPlus ---")
print(f"Revenue Total: ${total_revenue:,.2f}")
print(f"Costo Total Productos: ${total_cost_goods:,.2f}")
print(f"Inversión Marketing: ${total_marketing:,.2f}")
print(f"Profit Neto: ${net_profit:,.2f}")
print(f"¿Es rentable?: {'SÍ' if net_profit > 0 else 'NO'}")
```

```python
# 1. Ticket promedio por orden
ticket_promedio = orders_with_costs.groupby('id_pedido')['monto_total'].sum().mean()

# 2. Cantidad promedio por orden
cantidad_promedio = orders_with_costs.groupby('id_pedido')['cantidad'].sum().mean()

# 3. Producto más vendido (por cantidad)
top_producto = orders_with_costs.groupby('nombre_producto')['cantidad'].sum().idxmax()
top_producto_qty = orders_with_costs.groupby('nombre_producto')['cantidad'].sum().max()

# 4. Gasto por cada ID de Campaña
gasto_por_campana = marketing_clean.groupby('id_campaña')['gasto'].sum().sort_values(ascending=False)

print(f"--- Comportamiento de Ventas ---")
print(f"Ticket Promedio: ${ticket_promedio:.2f}")
print(f"Cantidad Promedio por Pedido: {cantidad_promedio:.2f}")
print(f"Producto Estrella: {top_producto} ({top_producto_qty} unidades)")
print(f"\n--- Gasto por Campaña  ---")
print(gasto_por_campana)
```

```python
# Explorar tabla events
# =========================
query_events = '''
SELECT *
FROM events;
'''
events = pd.read_sql(query_events, con=engine)
events.head()
```

```python
# 1. Agrupamos por nombre_evento y contamos usuarios únicos
query_funnel = events.groupby('nombre_evento')['id_usuario'].nunique().reset_index()
query_funnel.columns = ['nombre_evento', 'usuarios_unicos']

# 2. Definimos el orden lógico
orden_pasos = [
    'first_visit', 
    'select_item',      
    'add_to_cart', 
    'begin_checkout',   
    'add_payment_info', 
    'purchase'
]

# 3. Aplicamos el orden categórico
query_funnel['nombre_evento'] = pd.Categorical(query_funnel['nombre_evento'], categories=orden_pasos, ordered=True)

# 4. Ordenamos y eliminamos cualquier residuo que no esté en nuestra lista
query_funnel = query_funnel.dropna(subset=['nombre_evento']).sort_values('nombre_evento').reset_index(drop=True)

# 5. Cálculo de la Tasa de Conversión (Conversion Rate)
total_inicial = query_funnel['usuarios_unicos'].values[0]
query_funnel['conversion_rate'] = (query_funnel['usuarios_unicos'] / total_inicial) * 100

display(query_funnel)
```

```python
# PARTE 1: Totales del funnel
# ======================

query_totals = '''
SELECT 
    nombre_evento, 
    COUNT(DISTINCT id_usuario) as usuarios_unicos
FROM events
GROUP BY nombre_evento
ORDER BY 
    CASE 
        WHEN nombre_evento = 'first_visit' THEN 1
        WHEN nombre_evento = 'select_item' THEN 2
        WHEN nombre_evento = 'add_to_cart' THEN 3
        WHEN nombre_evento = 'begin_checkout' THEN 4
        WHEN nombre_evento = 'add_payment_info' THEN 5
        WHEN nombre_evento = 'purchase' THEN 6
        ELSE 7
    END;
'''

totals = pd.read_sql(query_totals, con=engine)
totals
```

```python
# PARTE 2: Conversiones
# ======================

query_conversion = '''
WITH base_usuarios AS (
    SELECT 
        nombre_evento, 
        COUNT(DISTINCT id_usuario) as usuarios_unicos
    FROM events
    GROUP BY nombre_evento
),
total_home AS (
    SELECT usuarios_unicos 
    FROM base_usuarios 
    WHERE nombre_evento = 'first_visit'
)
SELECT 
    b.nombre_evento,
    b.usuarios_unicos,
    ROUND((b.usuarios_unicos * 100.0) / (SELECT usuarios_unicos FROM total_home), 2) as conversion_rate
FROM base_usuarios b
ORDER BY 
    CASE 
        WHEN b.nombre_evento = 'first_visit' THEN 1
        WHEN b.nombre_evento = 'select_item' THEN 2
        WHEN b.nombre_evento = 'add_to_cart' THEN 3
        WHEN b.nombre_evento = 'begin_checkout' THEN 4
        WHEN b.nombre_evento = 'add_payment_info' THEN 5
        WHEN b.nombre_evento = 'purchase' THEN 6
        ELSE 7
    END;
'''

conversion = pd.read_sql(query_conversion, con=engine)
conversion
```

## 🔹 Paso 4: Evaluar si los usuarios regresan (retención por cohortes)

**🎯 Objetivo:** Analizar la retención de usuarios para entender si regresan después de registrarse y medir la fidelidad hacia RappiPlus.

**Tablas utilizadas:**
- `users`: Contiene la fecha de registro original de cada usuario.
- `user_activity`: Contiene el detalle de las interacciones posteriores al registro.

---

### 📑 Metodología de Análisis

1. **Identificación de Cohortes:** Se agrupó a los usuarios según su **mes de registro** (de enero a mayo de 2025) para observar su comportamiento a lo largo del tiempo.
2. **Cálculo de Retención Semanal:** Se midió cuántos usuarios únicos se mantuvieron activos en periodos de 7 días:
   - `retenido_w1`: Actividad entre los días 7 y 13.
   - `retenido_w2`: Actividad entre los días 14 y 20.
   - `retenido_w3`: Actividad entre los días 21 y 27.
3. **Métricas de Salud (Porcentajes):** Se calculó la efectividad de la retención dividiendo los usuarios activos entre el total inicial de la cohorte (`semana_1`, `semana_2`, `semana_3`).

> **Nota técnica:** Se aseguró la integridad de los datos normalizando las fechas con `CAST(fecha_registro AS DATE)` y utilizando `TO_CHAR` para una visualización limpia de los meses.

---

### 📉 Resultados y Conclusiones

Basado en el análisis de las cohortes de 2025, observamos los siguientes hallazgos:

* **Estabilidad del Producto:** ⚖️ La retención en la **Semana 1** se mantiene sólida por encima del **41%** en todas las cohortes, lo que indica una buena primera impresión del servicio.
* **Fidelización a Largo Plazo:** 💎 Sorprendentemente, la retención no presenta caídas drásticas hacia la **Semana 3**. De hecho, la cohorte de **febrero** muestra una retención del **43.98%** en su tercera semana, superando incluso su desempeño de la semana 1.
* **Conclusión General:** 🏆 RappiPlus presenta una retención sumamente saludable. Mantener tasas superiores al **40%** después de 21 días es un indicador claro de que los usuarios encuentran un valor recurrente en la suscripción, lo que reduce el costo de adquisición de clientes a largo plazo.

```python
# Explorar tabla users
# ======================
query_users = '''
SELECT 
    id_usuario, 
    CAST(fecha_registro AS DATE) as fecha_registro
FROM users;
'''

users = pd.read_sql(query_users, con=engine)
users.head(3)
```

```python
# Explorar tabla user_activity
# ======================
query_user_activity = '''
SELECT 
    id_usuario, 
    fecha_actividad, 
    dias_despues_registro, 
    activo
FROM user_activity;
'''

user_activity = pd.read_sql(query_user_activity, con=engine)
user_activity.head(3)
```

```python
# Retención por cohortes (Versión con formato de fecha limpio)
# ======================

query_cohort_retention_final = '''
WITH cohort_base AS (
    SELECT 
        id_usuario,
        -- Formateamos la fecha directamente en SQL para que solo muestre YYYY-MM
        TO_CHAR(DATE_TRUNC('month', CAST(fecha_registro AS DATE)), 'YYYY-MM') as mes_cohorte
    FROM users
),
retencion_calculada AS (
    SELECT 
        c.mes_cohorte,
        c.id_usuario,
        MAX(CASE WHEN a.dias_despues_registro BETWEEN 7 AND 13 AND a.activo = 1 THEN 1 ELSE 0 END) as w1,
        MAX(CASE WHEN a.dias_despues_registro BETWEEN 14 AND 20 AND a.activo = 1 THEN 1 ELSE 0 END) as w2,
        MAX(CASE WHEN a.dias_despues_registro BETWEEN 21 AND 27 AND a.activo = 1 THEN 1 ELSE 0 END) as w3
    FROM cohort_base c
    LEFT JOIN user_activity a ON c.id_usuario = a.id_usuario
    GROUP BY c.mes_cohorte, c.id_usuario
)
SELECT 
    mes_cohorte,
    COUNT(DISTINCT id_usuario) as usuarios_iniciales,
    COUNT(DISTINCT CASE WHEN w1 = 1 THEN id_usuario END) as retenido_w1,
    COUNT(DISTINCT CASE WHEN w2 = 1 THEN id_usuario END) as retenido_w2,
    COUNT(DISTINCT CASE WHEN w3 = 1 THEN id_usuario END) as retenido_w3,
    ROUND(COUNT(DISTINCT CASE WHEN w1 = 1 THEN id_usuario END) * 100.0 / COUNT(DISTINCT id_usuario), 2) as semana_1,
    ROUND(COUNT(DISTINCT CASE WHEN w2 = 1 THEN id_usuario END) * 100.0 / COUNT(DISTINCT id_usuario), 2) as semana_2,
    ROUND(COUNT(DISTINCT CASE WHEN w3 = 1 THEN id_usuario END) * 100.0 / COUNT(DISTINCT id_usuario), 2) as semana_3
FROM retencion_calculada
GROUP BY mes_cohorte
ORDER BY mes_cohorte;
'''

cohorte_final = pd.read_sql(query_cohort_retention_final, con=engine)
cohorte_final
```

## 🔹 Paso 5: Validar si los cambios generan impacto (test estadístico)

🎯 **Objetivo:** Evaluar si la modificación en la UI del checkout impacta la **tasa de conversión de compra**.

---

1. **Analizar el dataset** `experiment_checkout_ui.csv` para identificar la métrica principal **conversion**.
    - La métrica **conversion** es 1 si el usuario completó la compra, 0 si no.    
2. **Plantear la hipótesis estadística**
3. **Aplicar el test estadístico adecuado**
4. **Interpretar el resultado** ---
### Hipótesis estadística

- **$H_0$ (Hipótesis nula):** No existe una diferencia significativa entre las tasas de conversión del grupo de control y el de tratamiento.
- **$H_1$ (Hipótesis alternativa):** Existe una diferencia estadísticamente significativa entre las tasas de conversión de ambos grupos.
   
**Test estadístico:** Prueba Z para proporciones (Z-test for proportions).  
**Nivel de significancia alpha:** 0.05 (5%).

---

### Interpretación de resultados

Tras ejecutar el análisis con los datos del experimento, se obtuvieron los siguientes valores clave:

* **Estadístico Z:** 0.8133
* **P-value:** 0.4161

**Conclusión:** Dado que el **p-value (0.4161)** es mayor al nivel de significancia establecido (**0.05**), **no se rechaza la hipótesis nula**. Esto significa que no hay evidencia estadística suficiente para afirmar que el cambio en la UI del checkout haya generado un impacto real en la conversión de los usuarios. La diferencia observada entre el grupo de control (15.69%) y tratamiento (16.29%) se atribuye al azar.

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

# 1. Cargar el dataset
url = 'https://practicum-content.s3.amazonaws.com/datasets/experiment_checkout_ui.csv'
df_experiment = pd.read_csv(url)

# 2. Exploración rápida por variante
resumen = df_experiment.groupby('variante')['convirtio'].agg(['count', 'sum', 'mean'])
resumen.columns = ['usuarios_totales', 'conversiones', 'tasa_conversion']
display(resumen)
```

```python
# Éxitos (conversiones)
successes = [resumen.loc['tratamiento', 'conversiones'], resumen.loc['control', 'conversiones']]

# Observaciones (totales)
nobs = [resumen.loc['tratamiento', 'usuarios_totales'], resumen.loc['control', 'usuarios_totales']]

print(f"Éxitos (Tratamiento, Control): {successes}")
print(f"Muestra total (Tratamiento, Control): {nobs}")
```

```python
from statsmodels.stats.proportion import proportions_ztest

# Ejecutar el test
z_stat, p_value = proportions_ztest(successes, nobs)

print(f"Estadístico Z: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretación lógica
alpha = 0.05
if p_value < alpha:
    print("\n✅ CONCLUSIÓN: La diferencia es ESTADÍSTICAMENTE SIGNIFICATIVA.")
    print("El cambio en la UI impactó positivamente la conversión.")
else:
    print("\n❌ CONCLUSIÓN: La diferencia NO es estadísticamente significativa.")
    print("No hay evidencia suficiente para asegurar que el cambio de UI sea mejor que el actual.")
```
